```python
import pandas as pd
import matplotlib.pyplot as plt
cd=pd.read_csv("C://Users//CVR//Downloads//GlobalLandTemperaturesByCity.csv")
cd
```

```python
cd.isnull()
```

```python
cd.dropna()
```

```python
cd.dropna(inplace=True)
```

```python
cd.isnull().sum()
```

```python
import pandas as pd
cd['dt'] = pd.to_datetime(cd['dt'])
cd['year'] = cd['dt'].dt.year
mean_temp = cd.groupby('year')['AverageTemperature'].mean()
print(mean_temp)
plt.figure(figsize=(10, 6))
mean_temp.plot(kind='line')
plt.title('Average Temperature for Each Year')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.show()

```

```python
max_avg_temp=cd.groupby('City')['AverageTemperature'].max().head(10)
print(max_avg_temp)
```

```python
#group by country and calculate each average temperature
```

```python

mean_avg_temp = cd.groupby('Country')['AverageTemperature'].mean().head(10)
top_countries = mean_avg_temp.sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_countries.index, top_countries.values, color='orange')
plt.title('Top 10 Countries by Average Temperature')
plt.xlabel('Country')
plt.ylabel('Average Temperature (°C)')
plt.show()

```

```python
temp_data=cd['AverageTemperature'].dropna()
plt.figure(figsize=(10,6))
plt.hist(temp_data,bins=30,color='skyblue',edgecolor='black')
plt.title('Distribution of avg temp',fontsize=16)
plt.xlabel('Temperature (°C)',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
print(tips.head())
```

```python
sns.histplot(data=tips,x="total_bill",bins=20,color="skyblue",kde=True)
plt.title("Distribution of total bills")
plt.show()
```

```python
sns.kdeplot(data=tips,x="total_bill",shade=True,color="purple")
plt.title("KDE Plot")
plt.show()
```

```python
sns.boxplot(data=tips,x="day",y="total_bill",hue="sex",palette="coolwarm")
plt.title("Box plot of total bills   by day and gender")
plt.show()
```

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="time",palette="deep")
plt.title("Scatter plot of total bill vs tip")
plt.show()
```

```python
sns.lineplot(data=tips,x="size",y="tip",marker="o",color="green")
plt.title("Line plot of party size vs tip")
plt.show()
```

```python
sns.pairplot(data=tips,hue="sex",palette="husl")
plt.show()
```

```python
data = tips.select_dtypes(include=['number'])

cor_mat=data.corr()
sns.heatmap(cor_mat,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation matrix")
plt.show()
```
